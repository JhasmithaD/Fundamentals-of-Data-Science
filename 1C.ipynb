```python
import pandas as pd
structured_data = {'ID': [1, 2, 3],
                   'Name': ['Alice', 'Bob', 'Charlie'],
                   'Age': [30, 25, 35]}
df_structured = pd.DataFrame(structured_data)
print("Structured Data (DataFrame):")
display(df_structured)
```

```python
import json
semi_structured_data = {
    "employees": [
        {"id": 1, "name": "Alice", "skills": ["Python", "SQL"]},
        {"id": 2, "name": "Bob", "age": 25},
        {"id": 3, "name": "Charlie", "skills": ["Java", "C++"], "city": "New York"}
    ]
}
print("\nSemi-structured Data (JSON):")
print(json.dumps(semi_structured_data, indent=2))
```

```python
unstructured_data = "This is a sample text document. It contains free-form text without any specific structure. Analyzing this type of data requires specialized tools and techniques."
print("\nUnstructured Data (Text):")
print(unstructured_data)
```
