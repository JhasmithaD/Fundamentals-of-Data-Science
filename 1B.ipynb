```python
import pandas as pd
import matplotlib.pyplot as plt
data = {'job_title': ['Data Scientist', 'Data Analyst', 'Data Engineer', 'Data Scientist', 'Machine Learning Engineer',
                      'Data Analyst', 'Data Scientist', 'Data Engineer', 'Data Scientist', 'Data Analyst',
                      'Data Architect', 'Data Scientist', 'Data Analyst', 'Data Engineer', 'Data Scientist']}
df_roles = pd.DataFrame(data)
role_counts = df_roles['job_title'].value_counts()
print("Distribution of Data Science Roles:")
display(role_counts)
plt.figure(figsize=(12, 6))
role_counts.plot(kind='bar')
plt.title('Distribution of Data Science Roles')
plt.xlabel('Data Science Role')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
