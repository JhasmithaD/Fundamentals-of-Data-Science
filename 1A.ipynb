```python
import numpy as np
import pandas as pd
years = np.arange(2015, 2025)
job_postings = (years - 2014) * 5000 + np.random.randint(-2000, 2000, size=len(years))
job_postings[job_postings < 0] = 0
df_job_trends = pd.DataFrame({'year': years, 'num_postings': job_postings})
display(df_job_trends)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(df_job_trends['year'], df_job_trends['num_postings'])
plt.title('Trend of Data Science Job Postings (Last Decade)')
plt.xlabel('Year')
plt.ylabel('Number of Job Postings')
plt.grid(True)
plt.show()
```
